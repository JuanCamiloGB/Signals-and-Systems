# **PARA SISTEMAS RLC SERIE**

Dado que es un circuito RLC serie su factor de amortiguamiento esta dado por la siguiente ecuación:
$\zeta=\frac{R}{2\sqrt{\frac{L}{C}}}$

##**SISTEMA SUBAMORTIGUADO**

###Para que el sistema sea subamortiguado $0<ζ<1$ 

En este caso se tomaron los valores $R = 0.55 Ω , L = 0.6H, C = 0.5μF$. Para un $\zeta=0.5$ 
\
$$0< 0.5 < 1$$

```python
import sympy as sym
import matplotlib.pyplot as plt  
sym.init_printing()

S = sym.symbols('S', complex=True)
t, R, L , C = sym.symbols('t R L C', positive=True )
#Entrada Laplace 
V_i = 1/S #Señal de entrada
H_s = (1/L*C)/(S**2 + (R/L)*S + 1/(L*C))  #Funcion de transferencia
V_c = H_s*V_i #Señal de Salida
V_c
#Función de salida en Laplace factorizada (fracciones parciales)
Y_p = V_c.apart(S) # apart factoriza la expresión 
Y_p
#RLC en funcion del tiempo
Y_e = Y_p
#Modelo en el tiempo con RLC evaluados
RLC1 = {R:sym.Rational('.55') , L: sym.Rational('.6'), C: sym.Rational('2')}#valores utilizados para la simulación
Y_e = Y_p.subs(RLC1)
for j in range(len(Y_e.args)):
    if j==0:
      yt = sym.inverse_laplace_transform(Y_e.args[j], S, t)
    else:
      yt+= sym.inverse_laplace_transform(Y_e.args[j], S, t)

#Gráfica de salida en el tiempo
sym.plot(yt, (t, 0,17), xlabel='$t$', ylabel='$y(t)$',line_color='k',title="RLC Subamortiguado")
plt.show()
```

# **PARA SISTEMAS RLC PARALELO**
Dado que es un circuito RLC paralelo su factor de amortiguamiento esta dado por la siguiente ecuación: $\zeta=\frac{1}{2R\sqrt{\frac{C}{L}}}$

##**SISTEMA SUBAMORTIGUADO**

###Para que el sistema sea subamortiguado $0<ζ<1$ 

En este caso se tomaron los valores $R = 0.55 Ω , L = 0.6H, C = 0.5μF$. Para un $\zeta=0.5$
