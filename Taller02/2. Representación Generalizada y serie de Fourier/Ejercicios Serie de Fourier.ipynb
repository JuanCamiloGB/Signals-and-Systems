# a) $3t+4$

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
import math
from math import *
%matplotlib inline
import matplotlib as mpl
from ipywidgets import interact,IntSlider
import warnings
warnings.filterwarnings("ignore")
pi= np.pi
T=4
Fs = 500 # 500 muestras por segundo Ajustar según señal estudiada
Nm = T*Fs #numero de muestras requeridas
t = np.linspace(0,2,Nm)

wo= pi/2
N=5
m=2
cn = np.zeros(2*N+1,dtype=np.complex_)
nv = np.linspace(-N,N, 2*N+1) #vector num armónicos
Cn = 4*(-1)**nv
Cn[N] = 1/2

phin = np.zeros((int(Nm),2*N+1),dtype=np.complex_)
for n in range(-N,N+1,1):
    phin[:,n+N] = np.exp(1j*n*wo*t)

y= 3*t+4
xe=phin@Cn
fig = plt.figure(figsize=(18,4))
plt.subplot(131)
plt.plot(t,y)
plt.title(r'Señal Original')
plt.xlabel(r'$t(s)$',fontsize = 14)
plt.ylabel(r'$x(t)$',fontsize = 14)
plt.grid()

plt.subplot(132)
plt.stem(nv, np.real(Cn), 'r')
plt.title(r'Re{Cn}')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$Re\{C_n\}$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(133)
plt.stem(nv, np.imag(Cn), 'r')
plt.title(r'Im{Cn}')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$Im\{C_n\}$',fontsize = 14)
plt.axis('tight')
plt.grid()
plt.show()

fig = plt.figure(figsize=(18,4)) #magnitud 
plt.subplot(131)
plt.stem(nv, abs(Cn), 'r')
plt.title(r'|Cn|')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$|C_n|$',fontsize = 14)
plt.axis('tight')
plt.grid()


plt.subplot(132) 
plt.plot(t,y,label='$x(t)$')
plt.plot(t,xe,label='$\hat{x}(t)$')
plt.plot(t,y-xe,label='$e(t)$')
plt.axis('tight')
plt.grid()
plt.legend()
plt.show()

def pltest(Na=1):
    ind = range(N-Na,N+Na+1)
    er = 1 - (((4 + pi**2)/(pi**2)))
    xe = phin[:,ind]@ Cn[ind] 
    plt.plot(t,xe,color='b',label='xe')
    plt.plot(t,y,color='r',label='x')
    plt.title('Er=%.2f[%%]' % (100*er),fontsize=16)
    plt.xlabel('t[s]')
    plt.ylabel('x(t)')
    plt.grid()
    plt.legend()
    plt.show()
    return
interact(pltest,Na=(1,N,1))

fig.tight_layout() 
```
