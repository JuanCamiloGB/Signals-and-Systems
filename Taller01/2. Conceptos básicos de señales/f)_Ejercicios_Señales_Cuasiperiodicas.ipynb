1) 
$$x(t) = 3cos(ωt)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Solución"""
A= 3 
w= 300
t = np.arange(0,2*np.pi,0.25)
xGG = A*np.cos(w*t)
plt.figure(figsize=(12,4))
plt.plot(t,xGG,'k',linewidth=4,label="x(t)")
plt.ylabel("x(t)")
plt.title("Señal periodica")
plt.xlabel("t")
plt.grid()
plt.show()
```

2)
$$x(t) = 2sin(ωt+π)
$$

```python
"""Solución"""
A=2
w=300
t= np.arange(0,2*np.pi,0.25)
xGG = A*np.sin(w*t+np.pi)
plt.figure(figsize=(12,4))
plt.plot(t,xGG,'k',linewidth=4,label="x(t)")
plt.ylabel("x(t)")
plt.title("Señal periodica")
plt.xlabel("t")
plt.grid()
plt.show()
```
