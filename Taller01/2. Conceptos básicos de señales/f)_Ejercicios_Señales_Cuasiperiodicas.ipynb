1) 
$$x(t) = 3cos(ωt)
$$

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
"""Solución"""
A= 3 
w= 300
t = np.arange(0,2*np.pi,0.25)
xGG = A*np.cos(w*t)
plt.figure(figsize=(12,4))
plt.plot(t,xGG,'k',linewidth=4,label="x(t)")
plt.ylabel("x(t)")
plt.title("Señal periodica")
plt.xlabel("t")
plt.grid()
plt.show()
```

2)
$$x(t) = 2sin(ωt+π)
$$

```python
"""Solución"""
A=2
w=300
t= np.arange(0,2*np.pi,0.25)
xGG = A*np.sin(w*t+np.pi)
plt.figure(figsize=(12,4))
plt.plot(t,xGG,'k',linewidth=4,label="x(t)")
plt.ylabel("x(t)")
plt.title("Señal periodica")
plt.xlabel("t")
plt.grid()
plt.show()
```

3) $$x(t) = 3sin(√3t) +3sin(5t) − 2cos(t/√ 3)$$


```python
"""Solución"""
A1= 3
w1= np.sqrt(3)
A2= 3
w2= 5
A3= 2
w3= 1/np.sqrt(3)
t = np.arange(0,7,0.01)
xGG1=A1*np.sin(np.sqrt(3)*t)
xGG2=A2*np.sin(5*t)
xGG3=A3*np.cos((1/np.sqrt(3))*t)
xGG = xGG1+xGG2-xGG3
plt.figure(figsize=(15, 3))
plt.subplot(131)
plt.plot(t,xGG1,'b')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.subplot(132)
plt.plot(t,xGG2,'g')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.subplot(133)
plt.plot(t,xGG3,'r')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid
plt.figure(figsize=(15,3))
plt.plot(t,xGG,'k')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid
plt.show
```

4)$$x(t) = 3sin(4t) −2cos(50t) +2cos(10t)$$

```python
"""Solución"""
A1=  3
w1= 4
A2= 2
w2= 50
A3= 2
w3= 10
t= np.arange(0,7,0.01)
xGG1= A1*np.sin(w1*t)
xGG2= -A2*np.cos(w2*t)
xGG3= A3*np.cos(w3*t)
xGG= xGG1-xGG2+xGG3
plt.figure(figsize=(13,5))
plt.subplot(131)
plt.plot(t,xGG1,'b')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.subplot(132)
plt.plot(t,xGG2,'g')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.subplot(133)
plt.plot(t,xGG3,'r')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid
plt.figure(figsize=(15,3))
plt.plot(t,xGG,'k')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```

5)$$x(t) = e^{jωt}$$

```python
"""Solución"""
%matplotlib inline 
sym.init_printing()

t, omega = sym.symbols('t omega', real=True)
s = 1j*omega  
x = sym.exp(s*t)
y = x.subs({omega: 3}) #sustituir variables omega y sigma por valores numéricos
#se gráfica desde el paquete plot de sympy por ser objetos de variable simbólica
sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-1,1]) 
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-1,1])


```
