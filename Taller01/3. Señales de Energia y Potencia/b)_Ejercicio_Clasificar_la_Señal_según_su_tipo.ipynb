```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from numpy import matlib as mb
```

# Clasificar las señales segun su tipo (Energia o Potencia):

$$x(t)=3t + 2; \ \forall t\in [0; 5]$$

```python
t=np.arange(0,5,0.01)
xGG = 3*t+2
plt.plot(t,xGG,'g')
plt.grid()
plt.show
```

$$x(t)=Acos(\omega t)+Bsin(\omega t);  A,B,\omega  \in  \mathbb{R}^+$$ 


```python
A=5
B=5
w=10
t=np.arange(-5,5,0.01)
xGG = A*np.cos(w*t)+B*np.sin(w*t)
plt.plot(t,xGG,'g')
plt.grid()
plt.show()
```

$$x(t)=ate^{-tk} (u(t)-(u(t-t_o))  ;  a, k  \in  \mathbb{R} ; t_o>0$$ 

```python
from sympy import Symbol
t, w, A, B ,k, to=sym.symbols('t w A B k to',real=True)
t=Symbol('t')
step = sym.Heaviside(t)
sdeplace = sym.Heaviside(t-to)
x3= A*t*sym.exp(-k*t)*(step-sdeplace)
xt3 = x3.subs({A:1,k:2,to:2})
sym.plot(xt3,(t,-1,3),ylabel=r'x3(t)',line_color='r')
```

$$x(n)=nu[n]; \ n \in [0,\pm1,\pm2,...,\pm N]$$

```python
N = 15
t0 = 2
n= np.arange(-8, 10, 1)
len(n)
u = n >= t0 
xGG = n*u
plt.stem(n,xGG,use_line_collection=True)
plt.xlabel('n')
plt.grid(True)
```

$$x(n)=[n]; \ n \in [0,\pm1,\pm2,...,\pm N]$$

```python
N=6
n=np.arange(-N,N,6/10)
xGG=np.abs(n)
plt.stem(n, xGG, use_line_collection = True)
plt.grid()
plt.show()
```

$$x(n)=Acos[n\pi]u[n-n_0]; A \in \mathbb R^+ , n\in [0,\pm1,\pm2,...,\pm N];0<n_0<N$$

```python
N =15
n0=2
A = 3
n6=np.arange(-N,N,1)
u= n6 >= n0
xGG_6=A*np.cos(n6*np.pi)*u
plt.stem(n6,xGG_6,use_line_collection=True)
plt.grid()
plt.legend()
plt.xlabel('n')
plt.ylabel("$x[n]$")
plt.grid()
plt.show()
```
